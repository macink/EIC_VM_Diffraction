# Woods-Saxon density distribution functions in 2D

```c++
#include "WoodsSaxon_2D.h"
```

### Parameters for the functions below

```c++
double Vo = 2.12, R =6.38, a = 0.535;
double x_min = 0, x_max = 10, y_min = 0, y_max = 10, bins = 1000;
double tx_min = 0, tx_max = 0.1, ty_min = 0, ty_max = 0.1;

// Initialize constructor
WoodsSaxon_2D ws_init(Vo,R,a,x_min,x_max,y_min,y_max,bins,tx_min,tx_max,ty_min,ty_max);
```

### Woods-Saxon TF2

```c++
TF2 *ws_fun = ws_init.getWoodsSaxon2D();
    ws_fun->GetZaxis()->SetTitle("G(x,y) [fm^{-3}]");
    ws_fun->GetXaxis()->SetTitle("x [fm]");
    ws_fun->GetYaxis()->SetTitle("y [fm]");
    ws_fun->Draw();

c1->Draw();
```

### Woods-Saxon TH2D

```c++
TH2D *ws_hist = ws_init.getWoodsSaxonHist();
    ws_hist->GetZaxis()->SetTitle("G(x,y) [fm^{-3}]");
    ws_hist->GetXaxis()->SetTitle("x [fm]");
    ws_hist->GetYaxis()->SetTitle("y [fm]");
    ws_hist->Draw();

gStyle->SetOptStat(0);
c1->Draw();
```

```c++
TH2D *ws_hist_projX = ws_init.getWoodsSaxonHist();
    ws_hist_projX->GetZaxis()->SetTitle("G(x,y) [fm^{-3}]");
    ws_hist_projX->GetXaxis()->SetTitle("x [fm]");
    ws_hist_projX->GetYaxis()->SetTitle("y [fm]");
    ws_hist_projX->ProjectionX()->Draw();

gStyle->SetOptStat(0);
c1->Draw();
```

```c++
TH2D *ws_hist_projY = ws_init.getWoodsSaxonHist();
    ws_hist_projY->GetZaxis()->SetTitle("G(x,y) [fm^{-3}]");
    ws_hist_projY->GetXaxis()->SetTitle("x [fm]");
    ws_hist_projY->GetYaxis()->SetTitle("y [fm]");
    ws_hist_projY->ProjectionY()->Draw();

gStyle->SetOptStat(0);
c1->Draw();
```

### Woods-Saxon transform to momentum-space TF2

```c++
TF2 *ws_transform_fun = ws_init.getWStransform_fun();
    ws_transform_fun->GetZaxis()->SetTitle("|F(t_{x},t_{y})|^{2}");
    ws_transform_fun->GetXaxis()->SetTitle("t_{x} [GeV^{2}");
    ws_transform_fun->GetYaxis()->SetTitle("t_{y} [GeV^{2}");
    ws_transform_fun->Draw();

c1->Draw();
```

### Woods-Saxon transform to momentum-space TH2D

```c++
TH2D *ws_transform_hist = ws_init.getTransformed_hist();
    ws_transform_hist->GetZaxis()->SetTitle("|F(t_{x},t_{y})|^{2}");
    ws_transform_hist->GetXaxis()->SetTitle("t_{x} [GeV^{2}");
    ws_transform_hist->GetYaxis()->SetTitle("t_{y} [GeV^{2}");
    ws_transform_hist->Draw();

gStyle->SetOptStat(0);
c1->Draw();
```

```c++
TH2D *ws_transform_hist_projX = ws_init.getTransformed_hist();
    ws_transform_hist_projX->GetZaxis()->SetTitle("|F(t_{x},t_{y})|^{2}");
    ws_transform_hist_projX->GetXaxis()->SetTitle("t_{x} [GeV^{2}");
    ws_transform_hist_projX->GetYaxis()->SetTitle("t_{y} [GeV^{2}");
    ws_transform_hist_projX->ProjectionX()->Draw();

gPad->SetLogy(1);
gStyle->SetOptStat(0);
c1->Draw();
```

```c++
TH2D *ws_transform_hist_projY = ws_init.getTransformed_hist();
    ws_transform_hist_projY->GetZaxis()->SetTitle("|F(t_{x},t_{y})|^{2}");
    ws_transform_hist_projY->GetXaxis()->SetTitle("t_{x} [GeV^{2}");
    ws_transform_hist_projY->GetYaxis()->SetTitle("t_{y} [GeV^{2}");
    ws_transform_hist_projY->ProjectionY()->Draw();

gPad->SetLogy(1);
gStyle->SetOptStat(0);
c1->Draw();
```

# Woods-Saxon density distribution functions in 1D

```c++
#include "WoodsSaxon_1D.h"
```

### Parameters for the functions used below

```c++
double Vo = 2.12, R = 6.38, a = 0.535, r_min = 0, r_max = 10;
double bins = 1000, t_min = 0, t_max = 0.25;

// Initialize constructor
WoodsSaxon_1D ws_init(Vo,R,a,r_min,r_max,bins,t_min,t_max);
```

### Woods-Saxon TF1

```c++
TF1 *WS_fun = ws_init.getWoodsSaxon1D();
    WS_fun->GetYaxis()->SetTitle("G(r) [fm^{-3}]");
    WS_fun->GetXaxis()->SetTitle("r [fm]");
    WS_fun->Draw();

c1->Draw();
```

### Woods-Saxon TH1D

```c++
TH1D *ws_hist = ws_init.getWS_hist();
    ws_hist->GetYaxis()->SetTitle("G(r) [fm^{-3}]");
    ws_hist->GetXaxis()->SetTitle("r [fm]");
    ws_hist->Draw();

gStyle->SetOptStat(0);
c1->Draw();
```

### Woods-Saxon transform to momentum-space TF1

```c++
TF1 *ws_transform_fun = ws_init.getWStransform();
    ws_transform_fun->GetYaxis()->SetTitle("|F(t)|^{2}");
    ws_transform_fun->GetXaxis()->SetTitle("t [GeV^{2}]");
    ws_transform_fun->Draw();

gPad->SetLogy(1);
gStyle->SetOptStat(0);
c1->Draw();
```

### Woods-Saxon transform to momentum-space TH1D

```c++
TH1D *ws_transform_hist = ws_init.getWStransformHist();
    ws_transform_hist->GetYaxis()->SetTitle("|F(t)|^{2}");
    ws_transform_hist->GetXaxis()->SetTitle("t [GeV^{2}]");
    ws_transform_hist->Draw();

gPad->SetLogy(1);
gStyle->SetOptStat(0);
c1->Draw();
```

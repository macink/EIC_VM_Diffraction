# Form factor functions in 1D

```c++
#include "FormFactor_1D.h"
```

### These are the parameters for all functions below

```c++
double A = 197, Vo = 2.12, R = 6.38, a0 = 0.7;
double q_min = 0.001, q_max = 0.3, t_min = 0.0001, t_max = 0.1;

// Parameters to mess with
double phi_min = 0, phi_max = 2*pi;

// Initialize constructor
FormFactor_1D ff(A, Vo, R, a0, q_min, q_max, t_min, t_max, phi_min, phi_max);
```

## Form factor as a function of $q = \sqrt{t}$

```c++
TF1 *ff_q = ff.getFormFactorq_1D();
    ff_q->GetYaxis()->SetTitle("F(q)");
    ff_q->GetXaxis()->SetTitle("q [GeV]");
    ff_q->Draw();
```

```c++
c1->Draw();
```

```c++
TF1 *ff_q2 = ff.getFormFactorq2_1D();
    ff_q2->GetYaxis()->SetTitle("|F(q)|^{2}");
    ff_q2->GetXaxis()->SetTitle("q [GeV]");
    ff_q2->Draw();
```

```c++
c1->Draw();
```

## Form factor as a function of $t = q^{2}$

```c++
TF1 *ff_t = ff.getFormFactort_1D();
    ff_t->GetYaxis()->SetTitle("F(t)");
    ff_t->GetXaxis()->SetTitle("t [GeV^{2}]");
    ff_t->Draw();
```

```c++
c1->Draw();
```

```c++
TF1 *ff_t2 = ff.getFormFactort2_1D();
    ff_t2->GetYaxis()->SetTitle("|F(t)|^{2}");
    ff_t2->GetXaxis()->SetTitle("t [GeV^{2}]");
    ff_t2->Draw();
gPad->SetLogy(1);
```

```c++
c1->Draw();
```

```c++
TF1 *cut_ff_t2 = ff.getCutFormFactor_t2();
    cut_ff_t2->GetYaxis()->SetTitle("|F(t)|^{2}");
    cut_ff_t2->GetXaxis()->SetTitle("t [GeV^{2}]");
    cut_ff_t2->Draw();
```

```c++
c1->Draw();
```

## Compare true form factor with wedge cut

```c++
TF1 *ff_t2 = ff.getFormFactort2_1D();
    ff_t2->GetYaxis()->SetTitle("|F(t)|^{2}");
    ff_t2->GetXaxis()->SetTitle("t [GeV^{2}]");
    ff_t2->SetLineColor(kBlack);
    ff_t2->SetLineStyle(2);
    ff_t2->Draw();
TF1 *cut_ff_t2 = ff.getCutFormFactor_t2();
    cut_ff_t2->GetYaxis()->SetTitle("|F(t)|^{2}");
    cut_ff_t2->GetXaxis()->SetTitle("t [GeV^{2}]");
    cut_ff_t2->SetLineColor(kRed);
    cut_ff_t2->SetLineStyle(3);
    cut_ff_t2->Draw("same");
gPad->SetLogy(1);
```

```c++
c1->Draw();
```

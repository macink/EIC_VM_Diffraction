# Form factor functions in 2D: $|F(q_{x},q_{y})|^{2}$

```c++
#include "FormFactor_q_2D.h"
```

### These are the parameters for all functions below

```c++
double A = 197, Vo = 2.12, R = 6.38, a0 = 0.7;
double qy_min = 0, qy_max = 0.5, qx_min = 0, qx_max = 0.5;
double bins = 1000;

// Initialize constructor
FormFactor_q_2D ff_2d(A,Vo,R,a0,qy_min,qy_max,qx_min,qx_max,bins);
```

```c++
TH2D *ff_hist = ff_2d.getFormFactorq_2D_hist();
    ff_hist->SetTitle("2D Form Factor Contour: |F(qx,qy)|^{2}");
    ff_hist->GetYaxis()->SetTitle("q_{y} [GeV]");
    ff_hist->GetXaxis()->SetTitle("q_{x} [GeV]");
    ff_hist->GetZaxis()->SetTitle("|F(q)|^{2}");
    ff_hist->Draw();
gStyle->SetOptStat(0);
c1->Draw();
```

```c++
TH2D *ff_hist_projY = ff_2d.getFormFactorq_2D_hist();
    ff_hist_projY->SetTitle("2D Form Factor: |F(qx,qy)|^{2} y-Projection");
    ff_hist_projY->GetYaxis()->SetTitle("q_{y} [GeV]");
    ff_hist_projY->GetXaxis()->SetTitle("q_{x} [GeV]");
    ff_hist_projY->GetZaxis()->SetTitle("|F(q)|^{2}");
    ff_hist_projY->ProjectionY()->Draw();
gStyle->SetOptStat(0);
gPad->SetLogy(1);
c1->Draw();
```

```c++
TH2D *ff_hist_projX = ff_2d.getFormFactorq_2D_hist();
    ff_hist_projX->SetTitle("2D Form Factor: |F(qx,qy)|^{2} x-Projection");
    ff_hist_projX->GetYaxis()->SetTitle("q_{y} [GeV]");
    ff_hist_projX->GetXaxis()->SetTitle("q_{x} [GeV]");
    ff_hist_projX->GetZaxis()->SetTitle("|F(q)|^{2}");
    ff_hist_projX->ProjectionX()->Draw();
gStyle->SetOptStat(0);
gPad->SetLogy(1);
c1->Draw();
```

```c++
TF2* ff_fun = ff_2d.getFormFactorq2_2D();
    ff_fun->SetTitle("2D Form Factor TF2: |F(qx,qy)|^{2}");
    ff_fun->GetYaxis()->SetTitle("q_{y} [GeV]");
    ff_fun->GetXaxis()->SetTitle("q_{x} [GeV]");
    ff_fun->GetZaxis()->SetTitle("|F(q)|^{2}");
    ff_fun->Draw();
gStyle->SetOptStat(0);
gPad->SetLogy(0);
c1->Draw();
```

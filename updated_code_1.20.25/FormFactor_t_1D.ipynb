# Form factor functions in 1D: $|F(t)|^{2}$

```c++
#include "FormFactor_t_1D.h"
```

### These are the parameters for all functions below

```c++
double A = 197, Vo = 2.12, R = 6.38, a0 = 0.7;
double t_min = 0, t_max = 0.1, r_min = 0, r_max = 10;
double bins = 1000;

// Initialize constructor
FormFactor_t_1D ff(A,Vo,R,a0,t_min,t_max,bins,r_min,r_max);
```

### Form factor TF1

```c++
TF1 *ff_fun = ff.getFormFactort_1D();
    ff_fun->GetYaxis()->SetTitle("|F(t)|^{2}");
    ff_fun->GetXaxis()->SetTitle("t [GeV^{2}]");
    ff_fun->Draw();

gPad->SetLogy(1);
c1->Draw();
```

### Form factor TH1D

```c++
TH1D *ff_hist = ff.getFF_hist();
    ff_hist->GetYaxis()->SetTitle("|F(t)|^{2}");
    ff_hist->GetXaxis()->SetTitle("t [GeV^{2}]");
    ff_hist->Draw();

gStyle->SetOptStat(0);
gPad->SetLogy(1);
c1->Draw();
```

### Form factor transform to position space TF1

```c++
TF1 *ff_transformed_fun = ff.getTransformedFF_TF1();
    ff_transformed_fun->GetYaxis()->SetTitle("G(r) [fm^{-3}]");
    ff_transformed_fun->GetXaxis()->SetTitle("r [fm]");
    ff_transformed_fun->Draw();

gPad->SetLogy(0);
c1->Draw();
```

### Form factor transform to position space TH1D

```c++
TH1D *ff_transformed_hist = ff.getTransformedFF_hist();
    ff_transformed_hist->GetYaxis()->SetTitle("G(r) [fm^{-3}]");
    ff_transformed_hist->GetXaxis()->SetTitle("r [fm]");
    ff_transformed_hist->Draw();

gStyle->SetOptStat(0);
gPad->SetLogy(0);
c1->Draw();
```

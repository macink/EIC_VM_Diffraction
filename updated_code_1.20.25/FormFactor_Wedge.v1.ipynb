# Form factor Wedge Cuts

### These are the parameters for all functions below

```c++
double A = 197, Vo = 2.12, R = 6.38, a0 = 0.7, bins = 1000;
double t_min= 0, t_max = 0.1, r_min = 0, r_max = 10;

double phi_min = 0;
double phi_max1 = pi/2, phi_max2 = pi/3, phi_max3 = pi/4, phi_max4 = pi/6;
double phi_max5 = pi/9, phi_max6 = 2*pi, phi_max7 = pi;


// Initialize constructor
FormFactor_Wedge wedge1(A,Vo,R,a0,t_min,t_max,phi_min,phi_max1,bins);
FormFactor_Wedge wedge2(A,Vo,R,a0,t_min,t_max,phi_min,phi_max2,bins);
FormFactor_Wedge wedge3(A,Vo,R,a0,t_min,t_max,phi_min,phi_max3,bins);
FormFactor_Wedge wedge4(A,Vo,R,a0,t_min,t_max,phi_min,phi_max4,bins);
FormFactor_Wedge wedge5(A,Vo,R,a0,t_min,t_max,phi_min,phi_max5,bins);
FormFactor_Wedge wedge6(A,Vo,R,a0,t_min,t_max,phi_min,phi_max6,bins);
FormFactor_Wedge wedge7(A,Vo,R,a0,t_min,t_max,phi_min,phi_max7,bins);
FormFactor_t_1D ff_1d(A,Vo,R,a0,t_min,t_max,bins,r_min,r_max);
```

### Compare TF1's: wedge cut with true form factor (all cuts should be the same)

```c++
TF1 *trueFF = ff_1d.getFormFactort_1D();
    double trueFFIntegral = trueFF->Integral(trueFF->GetXmin(), trueFF->GetXmax());
    TF1 *normalizedTrueFF = new TF1("", [trueFF, trueFFIntegral](double *x, double *par) 
    {
        return trueFF->Eval(x[0]) / trueFFIntegral;
    }, trueFF->GetXmin(), trueFF->GetXmax(), 0);
    normalizedTrueFF->SetTitle("Compare |F(t)|^{2}");
    normalizedTrueFF->GetYaxis()->SetTitle("|F(t)|^{2}");
    normalizedTrueFF->GetXaxis()->SetTitle("t [GeV^{2}]");
    normalizedTrueFF->SetLineColor(kBlack);
    normalizedTrueFF->SetLineStyle(3);
    normalizedTrueFF->Draw();

TF1 *wedgeFF = wedge1.getCutFormFactor_t2_1D();
    double wedgeFFIntegral = wedgeFF->Integral(wedgeFF->GetXmin(), wedgeFF->GetXmax());
    TF1 *normalizedWedgeFF = new TF1("", [wedgeFF, wedgeFFIntegral](double *x, double *par) 
    {
        return wedgeFF->Eval(x[0]) / wedgeFFIntegral;
    }, wedgeFF->GetXmin(), wedgeFF->GetXmax(), 0);
    normalizedWedgeFF->SetLineColor(kRed);
    normalizedWedgeFF->SetLineStyle(2);
    normalizedWedgeFF->Draw("same");

TLegend *legend = new TLegend(0.75, 0.8, 0.9, 0.9);
    legend->AddEntry(normalizedTrueFF, "True Form Factor", "l");
    legend->AddEntry(normalizedWedgeFF, "Wedge Form Factor", "l");
    legend->Draw();

gPad->SetLogy(1);
c1->Draw();
```

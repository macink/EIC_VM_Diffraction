# Form factor functions in 2D: $|F(t_{x},t_{y})|^{2}$

```c++
#include "FormFactor_t_2D.h"
```

### These are the parameters for all functions below

```c++
double A = 197, Vo = 2.12, R = 6.38, a0 = 0.7;
double ty_min = 0, ty_max = 0.1, tx_min = 0, tx_max = 0.1, bins = 1000;
double x_min= 0, x_max = 10, y_min = 0, y_max = 10;

// Initialize constructor
FormFactor_t_2D ff(A,Vo,R,a0,tx_min,tx_max,ty_min,ty_max,bins,x_min,x_max,y_min,y_max);
```

### Form factor TF2

```c++
TF2 *ff_fun = ff.getFormFactort2_2D();
    ff_fun->GetZaxis()->SetTitle("|F(t_{x},t_{y})|^{2}");
    ff_fun->GetYaxis()->SetTitle("t_{y} [GeV^{2}]");
    ff_fun->GetXaxis()->SetTitle("t_{x} [GeV^{2}]");
    ff_fun->Draw();

c1->Draw();
```

### Form factor TH2D

```c++
TH2D *ff_hist = ff.getFormFactort_hist();
    ff_hist->GetZaxis()->SetTitle("|F(t_{x},t_{y})|^{2}");
    ff_hist->GetYaxis()->SetTitle("t_{y} [GeV^{2}]");
    ff_hist->GetXaxis()->SetTitle("t_{x} [GeV^{2}]");
    ff_hist->Draw();

gStyle->SetOptStat(0);
c1->Draw();
```

```c++
TH2D *ff_hist_projX = ff.getFormFactort_hist();
    ff_hist_projX->GetZaxis()->SetTitle("|F(t_{x},t_{y})|^{2}");
    ff_hist_projX->GetYaxis()->SetTitle("t_{y} [GeV^{2}]");
    ff_hist_projX->GetXaxis()->SetTitle("t_{x} [GeV^{2}]");
    ff_hist_projX->ProjectionX()->Draw();

gPad->SetLogy(1);
gStyle->SetOptStat(0);
c1->Draw();
```

```c++
TH2D *ff_hist_projY = ff.getFormFactort_hist();
    ff_hist_projY->GetZaxis()->SetTitle("|F(t_{x},t_{y})|^{2}");
    ff_hist_projY->GetYaxis()->SetTitle("t_{y} [GeV^{2}]");
    ff_hist_projY->GetXaxis()->SetTitle("t_{x} [GeV^{2}]");
    ff_hist_projY->ProjectionY()->Draw();

gPad->SetLogy(1);
gStyle->SetOptStat(0);
c1->Draw();
```

### Form factor transform to position space TF2

```c++
TF2 *ff_transformed_fun = ff.getTransformed_TF2();
    ff_transformed_fun->GetZaxis()->SetTitle("G(x,y) [fm^{-3}]");
    ff_transformed_fun->GetYaxis()->SetTitle("y [fm]");
    ff_transformed_fun->GetXaxis()->SetTitle("x [fm]");
    ff_transformed_fun->Draw();

gPad->SetLogy(0);
c1->Draw();
```

### Form factor transform to position space TH2D

```c++
TH2D *ff_transformed_hist = ff.getTransformed_hist();
    ff_transformed_hist->GetZaxis()->SetTitle("G(x,y) [fm^{-3}]");
    ff_transformed_hist->GetYaxis()->SetTitle("y [fm]");
    ff_transformed_hist->GetXaxis()->SetTitle("x [fm]");
    ff_transformed_hist->Draw();

gStyle->SetOptStat(0);
gPad->SetLogy(0);
c1->Draw();
```

```c++
TH2D *ff_transformed_hist_projX = ff.getTransformed_hist();
    ff_transformed_hist_projX->GetZaxis()->SetTitle("G(x,y) [fm^{-3}]");
    ff_transformed_hist_projX->GetYaxis()->SetTitle("y [fm]");
    ff_transformed_hist_projX->GetXaxis()->SetTitle("x [fm]");
    ff_transformed_hist_projX->ProjectionX()->Draw();

gStyle->SetOptStat(0);
gPad->SetLogy(0);
c1->Draw();
```

```c++
TH2D *ff_transformed_hist_projY = ff.getTransformed_hist();
    ff_transformed_hist_projY->GetZaxis()->SetTitle("G(x,y) [fm^{-3}]");
    ff_transformed_hist_projY->GetYaxis()->SetTitle("y [fm]");
    ff_transformed_hist_projY->GetXaxis()->SetTitle("x [fm]");
    ff_transformed_hist_projY->ProjectionY()->Draw();

gStyle->SetOptStat(0);
gPad->SetLogy(0);
c1->Draw();
```
